```python
cities = pd.read_csv('latlong_list.csv')
cities
```

```python
cities_list = []
for ea in range(0,len(cities)):
#     print(cities['name'][ea])
    cities_list.append({"name": cities['names'][ea], "lat": cities['latitude'][ea], "lon": cities['longitude'][ea]})
cities_list
```

```python
#hiking api - Method: getTrails
i=1
for ea in range(0,len(cities_list)):
#     print(cities_list[ea]['lat'])
#     print(cities_list[ea]['lon'])
    lat = cities_list[ea]['lat']
    lon = cities_list[ea]['lon']
    max_distance = 100
    max_results = 500
    url = f'https://www.hikingproject.com/data/get-trails?lat={lat}&lon={lon}&maxDistance={max_distance}&maxResults={max_results}&key={hike_api}'
    response = requests.get(url).json()
    response_df = pd.DataFrame(response['trails'])
    response_df['city']=cities_list[ea]['name']
    response_df.to_sql(name = 'trails', con = engine, if_exists='append', index=False)
    print(f"{i}/{len(cities)}")
    i=i+1
    
```

```python
#routes api - Method: getTrails
i=1
for ea in range(0,len(cities_list)):
#     print(cities_list[ea]['lat'])
#     print(cities_list[ea]['lon'])
    lat = cities_list[ea]['lat']
    lon = cities_list[ea]['lon']
    max_distance = 100
    max_results = 500
    min_diff = '5.5'
    max_diff = '5.10'
    url = f'https://www.mountainproject.com/data/get-routes-for-lat-lon?lat={lat}&lon={lon}&maxDistance={max_distance}&maxResults={max_results}&minDiff={min_diff}&maxDiff={max_diff}&key={mtn_api}'
    response = requests.get(url).json()
    response_df = pd.DataFrame(response['routes'])
    response_df['city'] = cities_list[ea]['name']
    response_df.to_sql(name = 'routes', con = engine, if_exists='append', index=False)
    print(f"{i}/{len(cities)}")
    i=i+1
    
```
